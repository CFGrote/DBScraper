# Database creation from gffs downloaded from pseudomonas.com

```python
from gffutils import create_db
```

```python
ifile="/home/grotec/repos/SBW25DB/data/pseudomonas.com/Pseudomonas_fluorescens_SBW25_116.gff"

db = create_db(ifile,
                   ":memory:",
                   id_spec="ID",
                   force=True,
                   gtf_gene_key="gene_id",
                   gtf_transcript_key="transcript_id",
                   verbose=True,
                   checklines=0,
                   merge_strategy='error',
                   _keep_tempfiles=True,
                   )
```
